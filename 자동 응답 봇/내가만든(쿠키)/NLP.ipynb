```python
from konlpy.tag import Komoran
komoran=Komoran()
import json
import pandas as pd
```

```python
with open('intents.json', encoding='utf-8') as f:
    data = json.load(f)
```

```python
df = pd.DataFrame(data['intents'][:])
df
```

```python
df['tag'].unique()
```

```python
dict_for_data = {'intent' : [],
                 'sentence' : [],
                 'QorA' : []
                 }
list_for_NLP = []
for tag in df['tag'].unique() :
    for sentence in df[df['tag']==tag]['patterns'].iloc[0] :
        dict_for_data['intent'].append(tag)
        dict_for_data['sentence'].append(sentence)
        dict_for_data['QorA'].append('Q')
        list_for_NLP.append(sentence)

    #for sentence in df[df['tag']==tag]['responses'].iloc[0] :
        #dict_for_data['intent'].append(tag)
        #dict_for_data['sentence'].append(sentence)
        #dict_for_data['QorA'].append('A')
        #list_for_NLP.append(sentence)
```

```python
df_end = pd.DataFrame(dict_for_data)
df_end
```

```python
komoran.morphs('안녕. 나는 하늘색과 딸기를 좋아해')
```

```python
NLP_data = []
for sentence in list_for_NLP :
    NLP_data.append(komoran.morphs(sentence))
```

```python
NLP_data
```

```python
df_end['NLP'] = NLP_data
df_end
```
